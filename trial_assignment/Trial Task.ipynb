## RESULTS:

```python
print('Sessions by student attendance')
fig, ax = plt.subplots(figsize=(20, 6))
plt.bar(x = sessions.index, height = all_attended['Session'].value_counts(), )
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
for bars in ax.containers:
    ax.bar_label(bars,padding=3);
```

```python
print('Sessions by student registrations')
fig, ax = plt.subplots(figsize=(20, 6))
plt.bar(x = names, height = nums, )
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
for bars in ax.containers:
    ax.bar_label(bars,padding=3);
```

```python
print('Sessions by average NPS')
fig, ax = plt.subplots(figsize=(20, 6))
plt.bar(x = session_names, height = exit_tickets.groupby(['Session'])['nps'].mean(), )
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
for bars in ax.containers:
    ax.bar_label(bars,padding=3);
```
